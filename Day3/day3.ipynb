Task 6: Slicing Mastery

```python
nums = [10, 20, 30, 40, 50]
print("First 3 elements")
print(nums[1:4])
print("Last 3 elements")
print(nums[-3: ])
print("Middle element")
print(nums[int(len(nums)/2)])
print("Every second element")
print(nums[ : : 2])
print("Reversed list")
print(nums[ : : -1])
```

Task 9: Extend vs Append

```python
a = [1,2]
b = [3,4]
a.append(100)
print(a)
a.append(b)
print(a)


```

```python
a = [1,2]
b = [3,4]
a.extend(b)
print(a)
```
