Count frequency of characters

```python
s = "aabca"
d ={}
for ch in s:
    if ch not in d:
        d[ch] = 1
    else :
        d[ch] +=1
print(d)            
```

Count frequency of numbers in list

```python
list = [1,2,2,3,1,1]

d ={}
for i  in list:
    if i not in d:
        d[i] = 1
    else :
        d[i] +=1
print(d)            
```

Most frequent character

```python
s = "hello world ,Python is great , pppppppppp"
frq ={}

for ch in s:
    frq[ch] = frq.get(ch,0) + 1

print(frq)
maximum = max(frq , key=frq.get)
print(maximum)
```

Merge two dictionaries

```python
d1 = {'a':1}
d2 = {'b':2}

merged = {**d1, **d2}
print(merged)
```

Count vowels

```python
s = "meghana"
vowels = "aeiou"

count = {}

for ch in s:
    if ch in vowels:
        count[ch] = count.get(ch, 0) + 1

print(count)
```

Common elements (set)

```python
a = [1,2,3]
b = [4,5,6,1,2,3]

common = list(set(a) & set(b))
print(common)
```

```python
s ="aabbccddeeff"
freq ={}

for ch in s:
    freq[ch] = freq.get(ch , 0)+1

for ch in s:
    if freq[ch] == 1:
        print(ch)
        break
```

Group by frequency

```python
nums = [1,1,2,2,2,3]
freq = {}

for n in nums:
    freq[n] = freq.get(n, 0) + 1

print(freq)
```
