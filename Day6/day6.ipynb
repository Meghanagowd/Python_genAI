Function to check even or odd

```python
def odd_even(num):
    if num%2 == 0 :
        print("even")
    else :
        print("odd")

odd_even(23)            
```

Function to count vowels in string

```python
def count_vowel(s):
    count = 0
    for i in s:
        if i in "aeiouAEIOU":
            count +=1
    return count


s = 'aeiou'
print(count_vowel(s))
```

Function to find maximum of 3 numbers

```python
def max_three(a,b,c):
    if a>b and a>c :
        print(f" {a} is greater")
    elif b>a and b>c :
        print(f" {b} is greater ") 
    else :
        print(f" {c} is greater")    

max_three(2,3,4) 
max_three(c=4,b=3,a=2)       
```

```python
def fact(n):
    fact = 1
    for i in range(1,n+1):
        fact = fact*i
    print(fact)    

fact(5)        
```

Function to count frequency of characters


```python
def count_freq(s):
    freq ={}
    for ch in s:
        freq[ch] = freq.get(ch,0)+1

    return freq

d = count_freq("meghana")
print(d)

  
```

Function to check palindrome

```python
def check_palindrome(s):
    rev = s[::-1]
    if rev == s:
        return True
    return False

print(check_palindrome("mam"))
```

Function using *args

```python
def add(b,*a):
    sum = b 
    for n in a:
        sum +=n
    return sum

print(add(2,4,5))
```

**kwargs

```python
def details(**data):
    for key,value in data.items():
        print(f"{key} : {value}")

details(name="Meghana",age=20,city="Bangalore")        
```
