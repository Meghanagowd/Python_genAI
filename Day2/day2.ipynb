Task 1: Input Trap

```python
a = input("Enter first number: ")
b = input("Enter second number: ")

print("Sum:", a + b)   # concatenation
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Sum:", a + b)
print("Difference:", a - b)
print("Product:", a * b)
```

Task 2: All Print Methods

```python
name = "Meghana"
marks = 98

# Concatenation
print("My name is " + name + " and my marks is " + str(marks))

# Comma
print("My name is", name, "and my marks is", marks)

# f-string (best)
print(f"My name is {name} and my marks is {marks}")
```

Task 4: Type Checker

Task 5: Division Practice

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Division (/):", a / b)
print("Floor Division (//):", a // b)
```

Task 6: Comparison Check

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("a > b:", a > b)
print("a == b:", a == b)
print("a != b:", a != b)
```

Task 7: Logical Operator

```python
age = int(input("Enter age: "))
has_id = input("Do you have ID (True/False): ")

has_id = has_id == "True"

if age >= 18 and has_id:
    print("Eligible")
else:
    print("Not Eligible")
```

Task 8: If-Else Basics

```python
num = int(input("Enter number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

Task 9: Nested If

```python
username = input("Enter username: ")
password = input("Enter password: ")

if username == "admin":
    if password == "1234":
        print("Login successful")
    else:
        print("Wrong password")
else:
    print("Invalid user")
```

Task 10: Grade System

```python
marks = int(input("Enter marks: "))

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```
