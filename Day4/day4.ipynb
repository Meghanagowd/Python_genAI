1. String Functions

```python
text = "hello world"

print(text.upper())
print(text.lower())
print(text.title())
print(text.capitalize())
```

2.Searching in String

```python
text = "abcdefg"
print(text.find("bcd"))  # 1 (returns index or -1 if not found)
print(text.index("de"))    #3 , error if not found
```

3. Count & Replace

```python
text = " aabbccaaddeeff"
print(text.count("a"))
```

```python
text = " hello world "
print(text.replace("hello","Python"))
```

4. Strip (Remove Spaces)

```python
text = "   hello  world    " 
print(text.strip())
print(text.lstrip())
print(text.rstrip())
```

5. Split & Join

```python
text = "hello world"
words = text.split()
print(words)
```

```python
text = "hello.world.all"
words = text.split(".")
print(words)
```

```python
words = ["good" ,"morning","everyone"]
print(" ".join(words))
```

6. Checking Functions

```python
text = "abc"

print(text.isalpha())   # True
print(text.isdigit())   # False
print(text.isalnum())   # True
```

⚠️ Note

7. For Loop (Range)

```python
for i in range(1,11):
    print(i)
```

```python
for i in range(1,11,2):
    print(i)
```

8. Table Program

```python
n = int(input("enter a number : "))

for i in range(1,11):
    print(n*i)
```

```python
n = int(input(" enter a number : "))

for i in range(0, n*10+1,n):
    print(i)
```

9. Looping Through String

```python
text = "Meghana"
for ch in text:
    print(ch)
```

```python
for i in range(len(text)):
    print(text[i])
```

10. Looping Through List

```python
my_list = [1, 2, 3, 4, 5]

for i in my_list:
    print(i)
```

```python
s = "abcefaab"

for i in range(len(s)):
    if s[i] == 'a':
        print(i)
```

12. Count Occurrences


```python
s = "ababac"
count = 0 

for ch in s:
    if ch == 'a':
        count += 1

print(count)        
```

13. User Input List

```python
my_list = []
for i in range(5):
    n = input()
    my_list.append(n)
print(my_list)    
```

14. 2D List Traversal

```python
my_list = [[1,2,3],[5,6],[9,11]]

for row in my_list :
    for val in row :
        print(val)
```

15. Basic While Loop

```python
i = 1

while i < 10:
    print(i)
    i += 1
```

16. Sum of Digits

```python
n = 534243

sum = 0
for digit in str(n):
    sum += int(digit)

print(sum)
```

 17. Reverse a Number

```python
n = 123

rev = str(n)[::-1]
print(rev)
```
