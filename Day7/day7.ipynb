Squares using loop

```python
n = [1,2,3,4,5]
result = []

for i in n:
    result.append(i*i)
print(result)    
```

Squares using list comprehension

```python
nums = [1, 2, 3, 4]

result = [n * n for n in nums]
print(result)
```

Even numbers

```python
nums = [1, 2, 3, 4, 5, 6]

evens = [n for n in nums if n % 2 == 0]
print(evens)
```

Uppercase names

```python
names = ["ram", "sita"]

upper_names = [name.upper() for name in names]
print(upper_names)
```

Even / Odd labeling

```python
nums = [1, 2, 3]

result = ["even" if n % 2 == 0 else "odd" for n in nums]
print(result)
```

Squares of even numbers

```python
nums = [1, 2, 3, 4]

result = [n * n for n in nums if n % 2 == 0]
print(result)
```

Names length > 4

```python
names = ["meghana", "ram", "sita", "python"]

result = [name for name in names if len(name) > 4]
print(result)
```

Dictionary (number → square)

```python
result = {x: x * x for x in range(4)}
print(result)
```

Dictionary even/odd

```python
nums = [1, 2, 3, 4]

result = {x: ("even" if x % 2 == 0 else "odd") for x in nums}
print(result)
```

Take 5 inputs

```python
nums = [int(input()) for _ in range(5)]
print(nums)
```

Remove duplicates

```python
nums = [1, 2, 2, 3]

result = list(set(nums))
print(result)
```

Flatten 2D list

```python
matrix = [[1,2,3],[4,5] ,[6,7,8]]

for i in matrix:
    for j in i:
        print(j)
```

Try/Except (division)

```python
try:
    n = int(input())
    print(10/n)
except ZeroDivisionError:
    print("cannot divide by zero")    
```

Handle invalid input

```python
try:
    n = int(input("Enter number: "))
except ValueError:
    print("Enter a valid number")
```

Full try-except structure

```python
try :
    n = int(input("enter number: "))
except ValueError:
    print("enter a valid number")
except ZeroDivisionError:
    print("enter other than zero")
else :
    print("successfull intake")
finally :
    print("DONE")                
```
